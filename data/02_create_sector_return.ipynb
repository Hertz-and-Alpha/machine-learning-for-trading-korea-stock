## 가격 데이터 불러오기

```python
with pd.HDFStore('assets.h5') as store:
    prices = store['finance_datareader/prices'][['adj_close']]
    kospi_stock = store['kospi/stocks']
```

```python
sector = kospi_stock[['sector']]
```

### 종목들에 Sector 열 추가

```python
prices = prices.join(sector)
```

```python
prices.dropna(inplace=True)
```

```python
unique_sector = prices['sector'].unique()
```

```python
sector_return = (prices[prices.sector == unique_sector[0]]['adj_close']
                     .unstack().resample('M')
                     .last().pct_change()
                     .mul(100)
                     .mean(axis=1)
                     .dropna()
                     .to_frame()
                     .rename(columns={0:unique_sector[0]})
                    )
```

```python
sector_return_list = []
for sector in unique_sector:
    sector_return = (prices[prices.sector == sector]['adj_close']
                     .unstack()
                     .resample('M')
                     .last()
                     .pct_change()
                     .mul(100)
                     .mean(axis=1)
                     .dropna()
                     .to_frame()
                     .rename(columns={0:sector})
                    )
    sector_return_list.append(sector_return)
```

```python
sector_data = pd.concat(sector_return_list, axis=1).fillna(0)
```

```python
sector_data.index = sector_data.index.to_period()
```
